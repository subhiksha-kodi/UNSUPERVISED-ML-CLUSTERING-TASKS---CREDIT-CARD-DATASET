```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('CC GENERAL.csv')
df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
df.dtypes
```

```python
df['CUST_ID']=df['CUST_ID'].str.split('C').str[1]
df['CUST_ID']=df['CUST_ID'].astype(int)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['MINIMUM_PAYMENTS']=df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median())
df['CREDIT_LIMIT']=df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median())
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
skip_cols = ['ONEOFF_PURCHASES_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'PRC_FULL_PAYMENT']

for i in df.columns:
  if i in skip_cols:
        continue 
  Q1=df[i].quantile(0.25)
  Q3=df[i].quantile(0.75)
  IQR=Q3-Q1
  lower=Q1-1.5*IQR
  upper=Q3+1.5*IQR
  outlier=((df[i]<lower) | df[i]>upper).sum()
  if outlier:
    df[i]=df[i].astype(float).clip(lower,upper)
  outlier=((df[i]<lower) | df[i]>upper).sum()
  print(f'{i}: {outlier} outliers')
```

```python
for i in df.columns:
  print(f'Skewness of {i}: {df[i].skew()}')
```

```python
for i in df.columns:
  for j in range(3):
    if (df[i].skew()>1):
      df[i]=np.log1p(df[i])
    if (df[i].skew()<-1):
      df[i]=np.log1p(df[i].max()+1-df[i])
  print(f'Skewness of {i}: {df[i].skew()}')
```

```python
df.to_csv('clustered_data.csv',index=False)
```
